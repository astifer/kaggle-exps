```python
import os
import PIL
from PIL import Image

for dirpath, dirnames, filenames in os.walk(path):
    prefix = 'classification-of-butterflies/train_butterflies/train_split/class_'

    # butterfly_classification/classification-of-butterflies/train_butterflies/train_split/class_32/2882.jpg
    if not dirpath.startswith(prefix):
        continue

    print(dirpath.split(prefix)[-1], filenames)

    for i in range(len(filenames)):
        img = Image.open(os.path.join(dirpath, filenames[i]))
        img.save(os.path.join(path, "all_data"))
```
